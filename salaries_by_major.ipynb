```python
import pandas as pd

df = pd.read_csv('sample_data/salaries_by_college_major.csv')

df.tail()
df.isna()


clean_df = df.dropna()
clean_df.tail()
clean_df.columns

# max_median_index = clean_df['Mid-Career Median Salary'].idxmax()
# print(f"max_median_index: {max_median_index}")
# print(clean_df['Mid-Career Median Salary'][max_median_index])


# max_start_index = clean_df['Starting Median Salary'].idxmax()
# print(f"max_start_index: {max_start_index}")
# print(clean_df['Starting Median Salary'][max_start_index])
```

```python
""" What college major has the highest mid-career salary? How much do graduates with this major earn? (Mid-career is defined as having 10+ years of experience)."""
max_mid_career_index = clean_df['Mid-Career Median Salary'].idxmax()
print(f"College Major with highest mid-carrer Salary: {clean_df['Undergraduate Major'][max_mid_career_index]}")
print(f"$: {clean_df['Mid-Career Median Salary'][max_mid_career_index]}")
```

```python
""" Which college major has the lowest starting salary and how much do graduates earn after university? """
min_median_salary = clean_df['Starting Median Salary'].idxmin()
print(f"Lowest earning starting college major: {clean_df['Undergraduate Major'][min_median_salary]}")
print(f"$: {clean_df['Starting Median Salary'][min_median_salary]}")
```

```python
""" Which college major has the lowest mid-career salary and how much can people expect to earn with this degree? """
min_mid_career_index = clean_df['Mid-Career Median Salary'].idxmin()
print(f"Lowest earning mid-carrer major: {clean_df['Undergraduate Major'][min_mid_career_index]}")
print(f"Max that can be earned with this degree: {clean_df['Mid-Career 10th Percentile Salary'][min_mid_career_index]}")
```

```python
#Calculating the difference between starting salaries and top salaries
clean_df['Mid-Career 90th Percentile Salary'].subtract(clean_df['Mid-Career 10th Percentile Salary'])
```

```python
#Adding the resultant df as a column to our table
spread_col = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
clean_df.insert(1, 'Spread', spread_col)
clean_df.head()
```

```python
#Sorting by the Lowest Spread
low_risk = clean_df.sort_values('Spread')
low_risk[['Undergraduate Major', 'Spread']].head()
```

```python
"""Using the .sort_values() method, can you find the degrees with the highest potential? Find the top 5 degrees with the highest values in the 90th percentile. """
highest_potentional = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
highest_potentional[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

```python
"""Also, find the degrees with the greatest spread in salaries. Which majors have the largest difference between high and low earners after graduation."""
high_risk = clean_df.sort_values('Spread', ascending=False)
high_risk[['Undergraduate Major', 'Spread']].head()
```

```python
#Grouping the columns that we have by business type
clean_df.groupby('Group').count()
```
